# Actividad 5: Procesamiento de conjuntos de datos en Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

Imprimir los valores de las primeras filas

```python
dat = pd.read_csv("robot_sensors.csv")
print(dat.head())
```

Limpiar los datos

1. Eliminar valores nulos

```python
dat_clean = dat.dropna().copy()
print(dat_clean)
```

Limpiar valores atipicos

```python
dat_clean = dat_clean[dat_clean['Tiempo (s)'] >= 0]
dat_clean = dat_clean[dat_clean['Distancia (cm)'] >= 0]
dat_clean = dat_clean[dat_clean['Velocidad (cm/s)'] >= 0]
dat_clean = dat_clean[dat_clean['Temperatura (°C)'] >= 0]
print(dat_clean)
```

Normalizar los datos numéricos con Min-Max Scaling

```python
valores_num = dat_clean[['Distancia (cm)', 'Velocidad (cm/s)', 'Temperatura (°C)']]
scaler = MinMaxScaler()
valores_num_norm = scaler.fit_transform(valores_num)
print(valores_num_norm)
```

Datos en una gráfca

```python
plt.plot(dat_clean['Tiempo (s)'], valores_num_norm[:, 0], label='Distancia (cm)', color='blue', linewidth=1.5)
plt.title('Evolución de Variables a lo largo del Tiempo', fontsize=14)
plt.xlabel('Tiempo (s)', fontsize=12)
plt.ylabel('Valor Normalizado (0-1)', fontsize=12)
plt.legend()
plt.grid(True, which='both', linestyle='--', alpha=0.5)
```

```python
plt.plot(dat_clean['Tiempo (s)'], valores_num_norm[:, 1], label='Velocidad (cm/s)', color='blue', linewidth=1.5)

plt.title('Visualización de Datos Normalizados')
plt.xlabel('Tiempo (s)')
plt.ylabel('Valor Escalado (0 a 1)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
plt.plot(dat_clean['Tiempo (s)'], valores_num_norm[:, 2], label='Temperatura (°C)', color='blue', linewidth=1.5)

plt.title('Visualización de Datos Normalizados')
plt.xlabel('Tiempo (s)')
plt.ylabel('Valor Escalado (0 a 1)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
